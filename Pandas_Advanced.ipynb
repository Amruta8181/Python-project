# GroupBy Operations

GroupBy allows us to group data based on a category and apply aggregation functions

This helps analysts summarize large datasets into meaningful insights

## Common Operations
- Total sales by region
- Average order value by product
- Quantity sold by category

```python
df.groupby('Region')['Sales'].sum()
```

```python
df.groupby('Product').agg({'Sales' : ['sum', 'mean'],'Quantity': ['mean']})
```

# Merge and Join operations

```python
customers = pd.DataFrame({
    'Customer': ['Amit', 'Rahul', 'Sneha', 'Priya','Karan'],
    'City' : ['Mumbai', 'Delhi', 'Pune','Bangalore', 'Chennai']
})
```

```python
customers
```

```python
orders = pd.DataFrame({
    'Customer': ['Amit', 'Rahul', 'Sneha', 'Raj', 'Neha'],
    'OrderValue': [50000, 20000, 30000, 48000, 27000]
})
```

```python
orders
```

```python
# INNER JOIN

pd.merge(customers, orders, on = 'Customer', how = 'inner')
```

```python
# LEFT JOIN

pd.merge(customers, orders, on='Customer', how = 'left')
```

# Melt Function

```python
monthly_sales = pd.DataFrame({
    'Product': ['Laptop', 'Phone'],
    'Jan': [20000, 15000],
    'Feb': [25000, 20000],
    'Mar': [30000, 35000]
})
```
