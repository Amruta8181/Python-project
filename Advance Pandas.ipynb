```python
# Q3 Customer Purchase Analysis
Identify the top 5 customers based on total sales.
Hint: Use GroupBy with aggregation.
```

```python
top_customers = (df.groupby('Customer')['Sales'].sum()
               .sort_values(ascending = False)
               .head(5))
print(top_customers)
```

```python
# Q4 Regional Sales Performance, Calculate the total sales generated by each region.
Then identify: Which region generated the highest revenue?
```

```python
regional_sales = (df.groupby('Region')['Sales'].sum()
                      .sort_values(ascending = False)
                      )
print(regional_sales)
```

```python
highest_region = regional_sales.idxmax()
highest_revenue = regional_sales.max()

print(f"Highest Revenue Region: {highest_region}")
print(f"Total Revenue: {highest_revenue}")
```

```python
# Q5 Product Performance Analysis, Calculate the total quantity sold for each product.
Which product appears to be the most popular?
```

```python
product_quantity = (df.groupby('Product')['Quantity'].sum()
    .sort_values(ascending = False))
                   
print(product_quantity)
```

```python
most_popular_product = product_quantity.idxmax()
highest_quantity = product_quantity.max()

print(f"Most Popular Product: {most_popular_product}")
print(f"Total Quantity Sold: {highest_quantity}")
```

```python
# Q7 Pivot Table Creation Create a pivot table showing:
Rows → Region
Columns → Product
Values → Total Sales
Which region sells the most laptops? 
```

```python
# Create pivot table
pivot_table = pd.pivot_table(
    df,
    values='Sales',
    index='Region',
    columns='Product',
    aggfunc='sum',
    fill_value=0
)
print(pivot_table)
```

```python
# Region with highest laptop sales
top_laptop_region = pivot_table['Laptop'].idxmax()
top_laptop_sales = pivot_table['Laptop'].max()

print(f"Region with highest Laptop sales: {top_laptop_region}")
print(f"Total Laptop sales: {top_laptop_sales}")
```

```python
# Q8 Profitability Analysis Calculate the average profit for each product category.
# Which category is most profitable? 
```

```python
avg_profit_category = (
    df.groupby('Category')['Profit']
      .mean()
      .sort_values(ascending=False)
)

print(avg_profit_category)
```

```python
most_profitable_category = avg_profit_category.idxmax()
highest_avg_profit = avg_profit_category.max()

print(f"Most Profitable Category: {most_profitable_category}")
print(f"Average Profit: {highest_avg_profit}")
```

```python
# Q9 Discount Impact Study, Analyze the relationship between discount and profit.
# Tasks: Calculate the average profit for each discount level
# Identify whether higher discounts lead to lower profit.
```

```python
discount_profit = (
    df.groupby('Discount')['Profit']
      .mean()
      .sort_values(ascending=False)
)
print(discount_profit)
```

```python
import matplotlib.pyplot as plt

discount_profit.sort_index().plot(kind='line', marker='o')

plt.title("Discount vs Average Profit")
plt.xlabel("Discount")
plt.ylabel("Average Profit")
plt.grid(True)
plt.show()
```

```python
filtered_df = df[df['Sales'] > 3000]

print(filtered_df)
```

```python
# Observation
# High-value transactions are common across all regions
# East, West, North, and South all have orders with Sales greater than ₹3,000.
# This indicates that strong revenue opportunities exist throughout the business.
# Electronics products dominate high-sales orders
# Products such as Phone, Laptop, Tablet, and Monitor appear frequently.
# Electronics seem to be the primary contributors to high-value sales.
# Some high-sales orders still have low profit or losses
# Example: Order 1004 has Sales of ₹3,933 but Profit of -₹71.66.
# This suggests discounts or costs may be affecting profitability.
# Large sales often generate strong profits
# Several orders with sales above ₹4,000–₹5,000 show profits in the hundreds or even thousands, indicating good margins on many high-value transactions.
```
