# **3. Introduction to NumPy**

NumPy stands for **Numerical Python**.

It is used for:

- Numerical computations
- Working with arrays and matrices
- Mathematical and statistical operations
- Linear algebra

### Key Features

- Fast computation
- Multi-dimensional arrays
- Broadcasting

# **Creating NumPy Arrays**

```python
import numpy as np

# Creating array from list
arr1 = np.array([1,2,3,4,5])

# Creating array of zeros
arr2 = np.zeros((3,3))

# Creating array of ones
arr3 = np.ones((2,4))

print("Array from list:",arr1)
print("Zero matrix:\n",arr2)
print("Ones matrix:\n",arr3)
```

## **Indexing and Slicing in NumPy**

NumPy arrays allow us to access specific elements or a range of elements.

### Indexing
Indexing is used to access a **single element** from an array.

### Slicing
Slicing is used to access **multiple elements or a portion of an array**.

Syntax:

array[start : end]

```python
import numpy as np

# Creating a NumPy array
arr = np.array([10, 20, 30, 40, 50])

# Indexing
print("First element:", arr[0])
print("Third element:", arr[2])

# Slicing
print("Elements from index 1 to 3:", arr[1:4])
```

## **Reshaping NumPy Arrays**

Sometimes data needs to be converted from one shape to another.

NumPy provides the **reshape()** function to change the dimensions of an array without changing the data.

For example, we can convert a **1D array into a 2D array**.

```python
import numpy as np

# Creating a 1D array
arr = np.array([1,2,3,4,5,6])

# Reshaping into 2 rows and 3 columns
reshaped = arr.reshape(2,3)

print("Original Array:")
print(arr)

print("Reshaped Array:")
print(reshaped)
```

# **Broadcasting in NumPy**

Broadcasting allows NumPy to perform operations on arrays of different shapes.

Instead of repeating values manually, NumPy automatically expands smaller arrays to match larger arrays.

This makes computations faster and more efficient.

```python
import numpy as np

# Creating an array
arr = np.array([1,2,3])

# Adding a scalar value
result = arr + 5

print("Original Array:", arr)
print("After Broadcasting:", result)
```

# **NumPy Array Operations**

```python
a = np.array([10,20,30])
b = np.array([1,2,3])

print("Addition:",a+b)
print("Subtraction:",a-b)
print("Multiplication:",a*b)
```

# **NumPy Statistical Functions**

```python
data = np.array([10,20,30,40,50])

print("Mean:",np.mean(data))
print("Standard Deviation:",np.std(data))
print("Sum:",np.sum(data))
```

# **Linear Algebra Operations in NumPy**

NumPy provides functions for performing **linear algebra operations** such as matrix multiplication and dot products.

These operations are widely used in **machine learning, statistics, and scientific computing**.

```python
import numpy as np

# Creating matrices
A = np.array([[1,2],
              [3,4]])

B = np.array([[5,6],
              [7,8]])

# Matrix multiplication
result = np.dot(A, B)

print("Matrix A:")
print(A)

print("Matrix B:")
print(B)

print("Dot Product:")
print(result)
```

# **Importance of Arrays in Data Analytics**

Arrays are one of the most important data structures used in **numerical computing and data analytics**.

An **array** is a collection of elements stored in a structured format where all elements usually have the **same data type**.

### Why Arrays Are Important

1. **Efficient Data Storage**
   - Arrays store large amounts of numerical data efficiently.

2. **Faster Computation**
   - Operations on arrays are much faster compared to traditional Python lists.

3. **Supports Mathematical Operations**
   - Arrays allow vectorized operations such as addition, subtraction, multiplication, and division.

4. **Foundation for Data Analytics Libraries**
   - Libraries like **NumPy, Pandas, TensorFlow, and Scikit-learn** use arrays internally.

5. **Easy Data Manipulation**
   - Arrays make it easy to perform slicing, indexing, reshaping, and filtering operations.

### **Example Use Cases**

Arrays are commonly used in:

- Machine Learning
- Data Analysis
- Image Processing
- Scientific Computing
